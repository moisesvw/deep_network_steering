```python
import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline
def print_steering_summary(data, plot=False):
    positive_angles = data[data['steering'] > 0]['steering']
    negative_angles = data[data['steering'] < 0]['steering']
    neutral_angles = data[data['steering'] == 0]['steering']
    print("Total", len(data))
    print("+ count", len(positive_angles))
    print("- count", len(negative_angles))
    print("N count", len(neutral_angles))
    print("Max:", data['steering'].max() )
    print("Min:", data['steering'].min() )
    print("Mean:", data['steering'].mean() )
    print("Median:", data['steering'].median() )
    print("Mode:", data['steering'].mode() )

    if plot:
        plt.hist(data['steering'], bins='auto')

    return (positive_angles, negative_angles, neutral_angles)
```
